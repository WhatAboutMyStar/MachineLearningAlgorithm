```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

### 加载手写数字数据集

```python
digits = load_digits()
x = digits.data
y = digits.target
```

```python
x.shape
```

```python
y.shape
```

```python
plt.imshow(x[0].reshape(8,8),cmap = plt.cm.gray)
plt.show()
```

```python
y[0]
```

```python
x[0]
```

### 分割训练集和测试集

```python
x_train , x_test , y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=666666)
```

### 将MNIST降维到2维

```python
pca = PCA(n_components=2)
pca.fit(x_train)
x_train_2 = pca.transform(x_train)
x_train_2.shape
```

```python
x_test_2 = pca.transform(x_test)
```

### 在主成分所解释的方差

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_
```

### 对PCA降维可视化

```python
for i in range(10):
    plt.scatter(x_train_2[y_train==i,0],x_train_2[y_train==i,1])
plt.show()
```

### 输入希望解释的方差

```python
pca = PCA(0.95) #希望解释百分之95的方差
pca.fit(x_train)
x_train_95 = pca.transform(x_train)
x_train_95.shape
```

```python
x_test_95 = pca.transform(x_test)
x_test_95.shape
```

### 对比性能
PCA不仅能降维，还能去噪，加快计算速度

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
%%time
knn_clf = KNeighborsClassifier()
knn_clf.fit(x_train,y_train)
knn_clf.score(x_train,y_train)
```

```python
%%timeit
knn_clf_2 = KNeighborsClassifier()
knn_clf_2.fit(x_train_2,y_train)
knn_clf_2.score(x_test_2,y_test)
```
